```python
# Import necessary libraries
import numpy as np
from sentence_transformers import SentenceTransformer, util

# Load pre-trained model
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

# Function to generate embeddings
def get_embeddings(sentences):
    return model.encode(sentences, convert_to_tensor=True)

# Sample sentences
sentences = ['This is a sentence.', 'This is another sentence.']

# Generate embeddings
embeddings = get_embeddings(sentences)
# Print embeddings
print(embeddings)
```

```python
# Define a function for semantic search
def semantic_search(query, embeddings):
    query_embedding = get_embeddings([query])
    cos_scores = util.pytorch_cos_sim(query_embedding, embeddings)[0]
    top_results = np.argpartition(-cos_scores, range(3))[0:3]
    return top_results

# Example query
query = 'Find similar sentence.'
results = semantic_search(query, embeddings)
print(results)
```

```python
# Evaluation metrics
def evaluate_retrieval(true_labels, predictions):
    true_positive = len(set(predictions) & set(true_labels))
    precision = true_positive / len(predictions) if predictions else 0
    recall = true_positive / len(true_labels) if true_labels else 0
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) else 0
    return precision, recall, f1

# Dummy true labels and predictions
true_labels = [0, 1]
predictions = [0, 1, 2]
precision, recall, f1 = evaluate_retrieval(true_labels, predictions)
print(f'Precision: {precision}, Recall: {recall}, F1 Score: {f1}')
```
