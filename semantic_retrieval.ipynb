This notebook demonstrates semantic retrieval using various embedding techniques including GloVe, Word2Vec, and Sentence-BERT. The sections cover dataset loading, preprocessing, embedding generation, similarity retrieval, evaluation, and error analysis.

```ipython
# Importing necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import pairwise_distances
import gensim
from sentence_transformers import SentenceTransformer

# Loading the dataset
data = pd.read_csv('path/to/your/dataset.csv')
data.head()
```

```ipython
# Preprocessing
def preprocess(text):
    # Add your preprocessing steps here
    return text.lower()

data['processed_text'] = data['text_column'].apply(preprocess)  # Update 'text_column' to your column name
```

```ipython
# Generating GloVe embeddings
glove_model = gensim.models.KeyedVectors.load_word2vec_format('path/to/glove/model', binary=False)
def get_glove_embedding(text):
    words = text.split()
    embedding = np.mean([glove_model[word] for word in words if word in glove_model], axis=0)
    return embedding

data['glove_embeddings'] = data['processed_text'].apply(get_glove_embedding)
```

```ipython
# Generating Word2Vec embeddings
word2vec_model = gensim.models.Word2Vec(sentences=data['processed_text'].str.split(), vector_size=100, window=5, min_count=1)
def get_word2vec_embedding(text):
    words = text.split()
    embedding = np.mean([word2vec_model.wv[word] for word in words if word in word2vec_model.wv], axis=0)
    return embedding

data['word2vec_embeddings'] = data['processed_text'].apply(get_word2vec_embedding)
```

```ipython
# Generating Sentence-BERT embeddings
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
data['sentence_bert_embeddings'] = model.encode(data['processed_text'].tolist(), show_progress_bar=True)
```
